```python
#Import the below packages in order to run the necessary code throughout the project.
import csv
import pandas as pd
```

```python
#Loading in the Coffee Exports file
#Start with an empty list that the data can be placed into later.
yearsList = []
countryData = []
#Open the excel file as a csv file.
with open('C:\\Users\cstew\OneDrive\Desktop\Syracuse\Spring-2023\IST-652\Project\Coffee_export.csv', 'r') as csvfile:
#Name the data that the csv reader brings in, ensuring you are specifying it is being pulled in from excel and that you want to use "," as a delimiter.
    coffee_exports = csv.reader(csvfile, dialect='excel', delimiter=',')
#I need to skip the first row for now so I can attach the data to the rows
    header = next(coffee_exports)  # Skip the header row
#I extracted the years from the header row and place them in the empty years list as I still need them later on
    yearsList = header[1:]  # Extract the years from the header row
#Now I went through an ensure the country names were established
    for line in coffee_exports:
        country = line[0]
#I then made sure the export values were assigned to the correct country and year.
#I converted export values to integers to that I could run mathmatical code if necessary.
        exports = [int(val) for val in line[1:]]
#Then I link the country names to each row.
        countryData.append([country] + exports)

# Create a DataFrame with country data where I set 'Country' as the index since it is what I will use to link datasets later on.
coffee_exports_df = pd.DataFrame(countryData, columns=['Country'] + yearsList)
coffee_exports_df = coffee_exports_df.set_index('Country')

# Add a year header back into the data frame.
coffee_exports_df.columns.name = 'Year'

#Then I printed the dataframe to ensure it worked. 
print(coffee_exports_df)
```

```python
#Loading in the Importers Coffee Consumption file
#Start with an empty list that the data can be placed into later.
yearsList = []
countryData = []
#Open the excel file as a csv file.
with open('C:\\Users\cstew\OneDrive\Desktop\Syracuse\Spring-2023\IST-652\Project\Coffee_importers_consumption.csv', 'r') as csvfile:
#Name the data that the csv reader brings in, ensuring you are specifying it is being pulled in from excel and that you want to use "," as a delimiter. 
    importer_consumption = csv.reader(csvfile, dialect='excel', delimiter=',')
#I need to skip the first row for now so I can attach the data to the rows
    header = next(importer_consumption)  # Skip the header row
#I extracted the years from the header row and place them in the empty years list as I still need them later on
    yearsList = header[1:]  # Extract the years from the header row
#Now I went through an ensure the country names were established
    for line in importer_consumption:
        country = line[0]
#I then made sure the consumption values were assigned to the correct country and year.
#I converted consumption values to integers to that I could run mathmatical code if necessary.
        importer_consumption = [int(val) for val in line[1:]]
#Then I link the country names and coffee type to each row.    
        countryData.append([country] + importer_consumption)

# Create a DataFrame with country data where I set 'Country' as the index since it is what I will use to link datasets later on.
importer_consumption_df = pd.DataFrame(countryData, columns=['Country'] + yearsList)
importer_consumption_df = importer_consumption_df.set_index('Country')  # Set 'Country' as the index

# Add a year header back into the data frame.
importer_consumption_df.columns.name = 'Year'

#Remove the last column containing the summed consumtion for each country so that it doesnt skew my later code.
importer_consumption_df = importer_consumption_df.iloc[:, :-1]

#Then I printed the dataframe to ensure it worked. 
print(importer_consumption_df)
```

```python
#Q1: Do countries consume more of the coffee they specialize in producing?
#Using coffee_exports_df, coffee_production and domestic_consumption_df

#I joined the data sets based on the common column 'Country' using an inner join
merged_df = coffee_production_df.join(domestic_consumption_df, how='inner', lsuffix='_production', rsuffix='_consumption')

#I compare the'Coffee Type' between the two data sets. When comparing two columns using the != operator, it returns a Boolean value of True if the values in the columns are not equal and False if the values are equal.
comparison = merged_df.index.get_level_values('Coffee Type Produced') != merged_df.index.get_level_values('Coffee Type Consumed')

#I printed the country and the comparison match result. I added <30 to the country placeholder in the format string. This specifies that the country name should be left-aligned with a width of 30 characters, providing a visually pleasing output.
#I had any "True" values print as a "Mismatch" and any "False" values print as a "Match"
for index, result in zip(merged_df.index, comparison):
    country = index[0]
    match_result = "Match" if not result else "Mismatch"
    print(f"Country: {country:<30} Comparison Result: {match_result}")
```

```python
#Q2: Is there a difference in coffee amounts between what is produced/imported versus consumed? 
#And is this amount equal to what is exported or re-exported?
#Using the coffee_imports_df, coffee_re_exports_df, importer_consumption_df, coffee_produced_df, coffee_exports_df and domestic_consumption_df

#Starting with the countries that produce the coffee
#Difference between produced and domestically consumed
#I selected the columns between the first and last years in the yearsList for both data sets
x_year = yearsList[0]
y_year = yearsList[-1]
coffee_production_cols = coffee_production_df.loc[:, x_year:y_year]
domestic_consumption_cols = domestic_consumption_df.loc[:, x_year:y_year]

#I found the difference in values between the two data sets
production_domestic_consumption_diff_df = coffee_production_cols.subtract(domestic_consumption_cols)

#I printed the output
print(production_domestic_consumption_diff_df.head())

#Remaining Coffee vs Exported Match
#I rename the year columns in production_domestic_consumption_diff_df to drop the year after "/" so that the data sets can match correctly.
new_columns = {}
for col in production_domestic_consumption_diff_df.columns:
    new_columns[col] = col.split("/")[0]  # Extract the year part before "/"
prod_dom_consumption_diff_df = production_domestic_consumption_diff_df.rename(columns=new_columns)

#I joined the data sets based on the common column 'Country'
merged_df_domestic = coffee_exports_df.join(prod_dom_consumption_diff_df, how='inner', lsuffix='_exports', rsuffix='_consumption')

#I compared the coffee amounts between the two data sets
comparison = merged_df_domestic.loc[:, '1990_exports':'2019_exports'].eq(
    merged_df_domestic.loc[:, '1990_consumption':'2019_consumption']
)

#Iterated over the rows and printed the comparison result
for index, result in zip(merged_df_domestic.index, comparison):
    country = index[0]
    match_result = "Match" if not result else "Mismatch"
    print(f"Country: {country:<30} Comparison Result: {match_result}")

#Now for countries who import and re-export coffee.
#Difference between imported and importers consumtion
#I select the columns between the first and last years in the yearsList for both data sets
x_year = yearsList[0]
y_year = yearsList[-1]
coffee_imports_cols = coffee_imports_df.loc[:, x_year:y_year]
importer_consumption_cols = importer_consumption_df.loc[:, x_year:y_year]

#I found the difference in values between the two data sets
importers_consumption_diff_df = coffee_imports_cols.subtract(importer_consumption_cols)

#I print the output
print(importers_consumption_diff_df.head())

#Remaining Coffee vs Re-exported Match
#I joined the data sets based on the common column 'Country'
merged_df_importers = coffee_re_exports_df.join(importers_consumption_diff_df, how='inner', lsuffix='_exports', rsuffix='_consumption')

#I compared the coffee amounts between the two data sets
comparison = merged_df_importers.loc[:, '1990_exports':'2019_exports'].eq(
    merged_df_importers.loc[:, '1990_consumption':'2019_consumption']
)

#I iterated over the rows and printed the comparison result
for country, result in zip(merged_df_importers.index.get_level_values('Country'), comparison.iterrows()):
    match_result = "Match" if result[1].all() else "Mismatch"
    print(f"Country: {country:<30} Comparison Result: {match_result}")
```

```python
#Q3: Is there a pattern off an increase or decrease in coffee consumption over the years?
#And does coffee investment follow this patter?
#Using domestic_consumption_df, importer_consumption_df and green_coffee_df

#First for Domestic coffee consumption
import matplotlib.pyplot as plt

#I calculated the total consumption for each year
total_domestic_consumption = domestic_consumption_df.sum()

#I create a larger figure for the data to fit the graph
plt.figure(figsize=(10, 6))

#I created a bar graph where year is the x-axis and consumption is the y-axis
plt.bar(total_domestic_consumption.index, total_domestic_consumption)
plt.xlabel('Year')
plt.ylabel('Coffee Consumption')
plt.title('Total Domestic Coffee Consumption per Year')
#I Add rotation and font size to make the bars fit better
plt.xticks(rotation=45, fontsize=8)
#And I added more x-axis space so everything was readable.
plt.subplots_adjust(bottom=0.2)
plt.show()
```
